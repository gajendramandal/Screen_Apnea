```python
import cv2
import numpy as np

video_path = "/home/gajendramandal/sleep_apnea/ALL__AP/AP_3/thermal-raw/Frag_7.avi"
cap = cv2.VideoCapture(video_path)
pixel_intensity_threshold = 10
prev_gray = None
frame_id = 0

pixel_diffs = []

while cap.isOpened():
    ret, frame = cap.read()
    if not ret or frame_id > 1000:  # Check first 200 frames for analysis
        break

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    if prev_gray is not None:
        diff = cv2.absdiff(prev_gray, gray)
        pixel_diff_count = np.count_nonzero(diff > pixel_intensity_threshold)
        pixel_diffs.append(pixel_diff_count)
        print(f"Frame {frame_id}: Pixel difference count = {pixel_diff_count}")

    prev_gray = gray
    frame_id += 1

cap.release()

median_value = np.mean(pixel_diffs)
print("Mean:", median_value)
```

```python
import cv2
import numpy as np
import os

AP=132
i=1

folder_path = f"/home/gajendramandal/sleep_apnea/ALL__AP/AP_{AP}/thermal-raw"
video_extensions = ['.mp4', '.avi', '.mov', '.mkv']
video_count = 0
for file in os.listdir(folder_path):
    if any(file.endswith(ext) for ext in video_extensions):
        video_count += 1

print(f"Total number of videos in the folder: {video_count}")

#creating the folder if it doesnt exist
folder_path =f"/home/gajendramandal/sleep_apnea/ALL__AP/AP_{AP}/AP_{AP}_frames@1000"
# Create the folder if it doesn't exist
if not os.path.exists(folder_path):
    os.makedirs(folder_path)
    print("Folder created âœ…")
else:
    print("Folder already exists ðŸš¨")
###################### my function ##########################################################

def my_function(i,AP):
        video_path=f"/home/gajendramandal/sleep_apnea/ALL__AP/AP_{AP}/thermal-raw/Frag_{i}.avi"
        cap = cv2.VideoCapture(video_path)
        
        if not os.path.exists(video_path):
            print(f"Video file not found: {video_path}")
            return

        cap = cv2.VideoCapture(video_path)
        
        if not cap.isOpened():
            print(f"Cannot open video ")
            return
        # Initialize variables
        prev_frame = None
        frame_count = 0
        unique_count = 0
        threshold = 1000 

        while cap.isOpened():
            ret, frame = cap.read()
            if not ret:
                break

            # Convert frame to grayscale for simplicity
            gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

            if prev_frame is not None:
                # Compute absolute difference between current and previous frame
                diff = cv2.absdiff(prev_frame, gray_frame)
                diff_count = np.count_nonzero(diff > 10)  

                # If difference exceeds the threshold, consider it a unique frame
                if diff_count > threshold:
                    unique_count += 1
                    cv2.imwrite(f"/home/gajendramandal/sleep_apnea/ALL__AP/AP_{AP}/AP_{AP}_frames@1000/AP_{AP}_Feag_{i}_frame_{unique_count}.png", frame)

            # Update previous frame
            prev_frame = gray_frame
            frame_count += 1

        cap.release()
        print(f"Processed {frame_count} frames and saved {unique_count} unique frames.--frag_{i}")
                
for _ in range(video_count):
    my_function(i,AP)
    i=i+1
    
    
```
