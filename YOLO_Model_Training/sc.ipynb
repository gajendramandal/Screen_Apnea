```python
import os
import shutil
import random

# Define paths for images and labels
image_folder = r"E:\IITGN\yolo_model_training\images\train"  # Path to images
label_folder = r"E:\IITGN\yolo_model_training\labels\train"  # Path to labels

# Define output directories
output_base = r"E:\IITGN\yolo_model_training"  # Base output folder
output_train_images = os.path.join(output_base, "train", "images")
output_train_labels = os.path.join(output_base, "train", "labels")
output_test_images = os.path.join(output_base, "test", "images")
output_test_labels = os.path.join(output_base, "test", "labels")

# Create directories if they don't exist
for folder in [output_train_images, output_train_labels, output_test_images, output_test_labels]:
    os.makedirs(folder, exist_ok=True)

# Get all image filenames (without extensions)
image_files = [f for f in os.listdir(image_folder) if f.endswith(('.jpg', '.png', '.jpeg'))]
random.shuffle(image_files)  # Shuffle dataset

# Split dataset (80% train, 20% test)
split_ratio = 0.8
split_index = int(len(image_files) * split_ratio)
train_files = image_files[:split_index]
test_files = image_files[split_index:]

# Function to move images & labels
def move_files(image_list, src_img_folder, src_lbl_folder, dest_img_folder, dest_lbl_folder):
    for img in image_list:
        img_name = os.path.splitext(img)[0]  # Get image name without extension
        img_path = os.path.join(src_img_folder, img)
        label_path = os.path.join(src_lbl_folder, img_name + ".txt")  # Corresponding label
        
        # Move image
        shutil.move(img_path, os.path.join(dest_img_folder, img))

        # Move label if it exists
        if os.path.exists(label_path):
            shutil.move(label_path, os.path.join(dest_lbl_folder, img_name + ".txt"))

# Move train & test files
move_files(train_files, image_folder, label_folder, output_train_images, output_train_labels)
move_files(test_files, image_folder, label_folder, output_test_images, output_test_labels)

# Print summary
print(f"✅ Train set: {len(train_files)} images")
print(f"✅ Test set: {len(test_files)} images")
```
